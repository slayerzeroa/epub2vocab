```python
import os
import requests
from dotenv import load_dotenv

# 1) .env 로드
load_dotenv('C:/Users/slaye/VscodeProjects/epub2vocab/src/functions/connect_dictionary/src/.env')
api_key = os.getenv("DICTIONARY_KEY")
if not api_key:
    raise RuntimeError("DICTIONARY_KEY is not set in .env")

# 2) 단어 입력
word = input("Enter a word to look up: ").strip()

# 3) API 호출
url = f"https://www.dictionaryapi.com/api/v3/references/collegiate/json/{word}?key={api_key}"
print(f"Request URL: {url}")  # 디버깅용

resp = requests.get(url, timeout=8)  # 8초 타임아웃
resp.raise_for_status()

# 4) JSON 파싱
data = resp.json()
if not isinstance(data, list) or len(data) == 0:
    print(f"No definitions found for {word}")
    exit()

# 5) shortdef 출력
for i, entry in enumerate(data, 1):
    if "shortdef" in entry:
        print(f"---- Definition Set {i} ----")
        for d in entry["shortdef"]:
            print("-", d)
```
